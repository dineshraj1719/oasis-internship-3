```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

plt.style.use('ggplot')
```

```python
df = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")

df.head()
```

```python
print("Shape:", df.shape)

df.info()
```

```python
df.columns = df.columns.str.strip()

df.columns
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(
    df['Date'],
    dayfirst=True
)

df.info()
```

```python
df['Month'] = df['Date'].dt.month

df['Month_Name'] = df['Date'].dt.strftime('%B')
```

```python
numeric_df = df.select_dtypes(include=['float64','int64'])

plt.figure(figsize=(10,6))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.distplot(
    df['Estimated Unemployment Rate (%)']
)

plt.title("Unemployment Rate Distribution")

plt.show()
```

```python
state_unemployment = (
    df.groupby('Region')
    ['Estimated Unemployment Rate (%)']
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(12,6))

sns.barplot(
    x=state_unemployment.values,
    y=state_unemployment.index
)

plt.title(
    "Top 10 States with Highest Unemployment"
)

plt.xlabel("Average Unemployment Rate")

plt.show()
```

```python
state_low = (
    df.groupby('Region')
    ['Estimated Unemployment Rate (%)']
    .mean()
    .sort_values()
    .head(10)
)

plt.figure(figsize=(12,6))

sns.barplot(
    x=state_low.values,
    y=state_low.index
)

plt.title(
    "Top 10 States with Lowest Unemployment"
)

plt.show()
```

```python
monthly = (
    df.groupby('Month_Name')
    ['Estimated Unemployment Rate (%)']
    .mean()
)

monthly = monthly.reindex([
    'January',
    'February',
    'March',
    'April',
    'May',
    'June',
    'July',
    'August',
    'September',
    'October',
    'November'
])

plt.figure(figsize=(12,5))

monthly.plot(marker='o')

plt.ylabel("Unemployment Rate")

plt.title(
    "Monthly Unemployment Trend During COVID-19"
)

plt.show()
```

```python
monthly_emp = (
    df.groupby('Month_Name')
    ['Estimated Employed']
    .mean()
)

monthly_emp = monthly_emp.reindex([
    'January',
    'February',
    'March',
    'April',
    'May',
    'June',
    'July',
    'August',
    'September',
    'October',
    'November'
])

plt.figure(figsize=(12,5))

monthly_emp.plot(marker='o')

plt.title(
    "Average Employment Trend"
)

plt.ylabel("Employment")

plt.show()
```

```python
monthly_lpr = (
    df.groupby('Month_Name')
    ['Estimated Labour Participation Rate (%)']
    .mean()
)

monthly_lpr = monthly_lpr.reindex([
    'January',
    'February',
    'March',
    'April',
    'May',
    'June',
    'July',
    'August',
    'September',
    'October',
    'November'
])

plt.figure(figsize=(12,5))

monthly_lpr.plot(marker='o')

plt.title(
    "Labour Participation Trend"
)

plt.ylabel("Participation Rate")

plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.boxplot(
    x='Region.1',
    y='Estimated Unemployment Rate (%)',
    data=df
)

plt.title(
    "Regional Unemployment Analysis"
)

plt.xticks(rotation=45)

plt.show()
```

```python
pivot = df.pivot_table(
    values='Estimated Unemployment Rate (%)',
    index='Region',
    columns='Month_Name',
    aggfunc='mean'
)

plt.figure(figsize=(15,10))

sns.heatmap(
    pivot,
    cmap='Reds'
)

plt.title(
    "State-wise Unemployment Heatmap"
)

plt.show()
```

```python
top_states = (
    df.groupby('Region')
    ['Estimated Unemployment Rate (%)']
    .mean()
    .sort_values(ascending=False)
    .head(15)
)

top_states
```

```python
before_covid = df[
    df['Month'] <= 3
]['Estimated Unemployment Rate (%)'].mean()

after_covid = df[
    df['Month'] >= 4
]['Estimated Unemployment Rate (%)'].mean()

print("Before COVID:", before_covid)
print("After COVID:", after_covid)
```

```python
covid_df = pd.DataFrame({
    'Period':['Before COVID',
              'After COVID'],
    'Rate':[before_covid,
            after_covid]
})

sns.barplot(
    x='Period',
    y='Rate',
    data=covid_df
)

plt.title(
    "COVID-19 Impact on Unemployment"
)

plt.show()
```

```python
print("Highest Average Unemployment State:")
print(
    df.groupby('Region')
    ['Estimated Unemployment Rate (%)']
    .mean()
    .idxmax()
)

print()

print("Lowest Average Unemployment State:")
print(
    df.groupby('Region')
    ['Estimated Unemployment Rate (%)']
    .mean()
    .idxmin()
)
```
